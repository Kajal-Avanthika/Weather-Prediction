### Step3: Fitting the model

```python
# Pulling every row back and creating a target column
# Predicting tomorrow's maximum temperature
core_weather["target"] = core_weather.shift(-1)["temp_max"]
core_weather
```

```python
# remove the last row
core_weather = core_weather.drop("2024-10-31")
```

```python
from sklearn.linear_model import Ridge

reg = Ridge(alpha = 0.1)
features = ["precip","temp_max","temp_min"]
```

```python
core_weather.index.year.unique()
```

```python
# Training and Testing data for time series
# Predicting future from past data
train = core_weather[:"2023-12-31"]
train
```

```python
test = core_weather["2024-01-01":]
test
```

```python
reg.fit(train[features],train["target"])
```

```python
predictions = reg.predict(test[features])
predictions
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(test["target"],predictions)
```

```python
predictions.shape, test["target"].shape
```

```python
# On an averge we are 3.85 degrees off from the actual value
```

### Step4: Evaluating our model

```python
combined = pd.concat([test["target"],pd.Series(predictions,index = test.index)],axis = 1)
combined.columns = ["Actual","Predicted"]
combined
```

```python
combined.plot()
```

```python
# Checking the regression coefficients
reg.coef_ # The weights for the three features we gave in
```

```python
# Creating a function to make predictions easier
def weather_predictor(features,core_weather,reg): # features,dataframe,model
  train = core_weather[:"2023-12-31"]
  test = core_weather["2024-01-01":]
  reg.fit(train[features],train["target"])
  predictions = reg.predict(test[features])
  error = mean_absolute_error(test["target"],predictions)
  combined = pd.concat([test["target"], pd.Series(predictions,index = test.index)], axis = 1)
  return error,combined
```

```python
weather_predictor(features,core_weather,reg)
```

#### Adding rolling means

```python
core_weather["month_max"] = core_weather["temp_max"].rolling(30).mean()
core_weather
```

```python
# Adding some features
core_weather["month_day_max"] = core_weather["month_max"]/core_weather["temp_max"]
core_weather["max_min"] = core_weather["temp_max"]/core_weather["temp_min"]
features = ["precip","temp_max","temp_min","month_max","month_day_max","max_min"]
```

```python
core_weather = core_weather.drop(core_weather.index[:30])
core_weather
```

```python
# Passing to the function
error, combined = weather_predictor(features,core_weather,reg)
error,combined # error reduced slightly
```

```python
combined.plot()
```

```python
# Adding more features
# The monthly average depends on the average of days of that month only
monthly_average = core_weather["temp_max"].groupby(core_weather.index.month).apply(lambda x: x.expanding(1).mean())
monthly_average.reset_index(level = 0,drop = True,inplace = True)
core_weather["monthly_average"] = monthly_average
core_weather
```

```python
features = ["precip","temp_max","temp_min","month_max","month_day_max","max_min","monthly_average"]
error, combined = weather_predictor(features,core_weather,reg)
error,combined # Error reduced further
```
